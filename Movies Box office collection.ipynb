**Combining all three datasets into one**

**Total has all the data from 2000 to 2024 **

```python
total = pd.concat([df, ten,now], ignore_index=True)
total.sort_index()
total.head(5)
```

```python
total.dtypes
```

```python
total["Domestic"]= total['Domestic'].str.replace(",", "").astype(float)
total.head(5)
```

```python
total.dtypes
total["Worldwide"]= total['Worldwide'].str.replace(",", "").astype(float)
total.head(5)
```

```python
total["Foreign"]= total['Foreign'].str.replace(",", "").astype(float)
total.head(5)
```

```python
total['Domestic_percent'] = pd.to_numeric(total['Domestic_percent'].astype(str).str.replace('%', '').str.replace('<', ""), errors='coerce') / 100
```

```python
total['Foreign_percent'] = pd.to_numeric(total['Foreign_percent'].astype(str).str.replace('%', '').str.replace('<', ""), errors='coerce' )/ 100
```

**Visulization of the given data**

```python
bar =total.sort_values(by="Worldwide",ascending=False).head(10)\
   .plot(kind="bar",x="Name",y="Worldwide",title="Top 10 highest grossing Movies",color="green",ylim=(0,3000000000))
bar.set_ylabel('Worldwide Gross ($)')
bar.set_xlabel('Movie Name')
plt.ticklabel_format(style='plain', axis='y')
bar.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python

bar = total.sort_values(by="Worldwide", ascending=False).head(10)

# Plotting
ax = bar.plot(kind="bar", x="Name", y="Worldwide", color="blue", title="Top 10 Highest Grossing Movies")

# Customizing the plot
ax.set_ylabel('Worldwide Gross ($)')
ax.set_xlabel('Movie Name')
ax.set_ylim(0, 3_000_000_000)
ax.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Format y-axis
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

```python
ax= sns.scatterplot(x="Domestic",y="Foreign",hue="year",data=total)
ax.ticklabel_format(style="plain",axis="x")
```

```python
ab = total.groupby("year")["Worldwide"]\
    .agg(['count', 'mean'])\
    .sort_values(by='mean', ascending=True)\
    .head(10)

# Sort the index (years) to have a sequential x-axis
ab = ab.sort_index()

# Set the style and palette
sns.set(style="whitegrid")
plt.figure(figsize=(12, 12))

# Plot the line graph
sns.lineplot(x=ab.index, y=ab['mean'], marker='o', linewidth=4, color='darkblue')

# Title and labels with formatting
plt.title("Top 10 Years with the Highest Average Worldwide Collection", fontsize=18, weight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Worldwide Collection (in billions)", fontsize=14)

# Remove scientific notation from y-axis
plt.ticklabel_format(style="plain", axis='y')

# Add data labels for each point
for x, y in zip(ab.index, ab['mean']):
    plt.text(x, y, f'{y:,.2f}', ha='center', va='bottom', fontsize=12, weight='bold')

# Display the plot
plt.show()
```
