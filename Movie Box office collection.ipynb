```python
ax= sns.scatterplot(x="Domestic",y="Foreign",hue="year",data=total)
ax.ticklabel_format(style="plain",axis="x")
```

```python
ab = total.groupby("year")["Worldwide"]\
    .agg(['count', 'mean'])\
    .sort_values(by='mean', ascending=True)\
    .head(10)

# Sort the index (years) to have a sequential x-axis
ab = ab.sort_index()

# Set the style and palette
sns.set(style="whitegrid")
plt.figure(figsize=(12, 12))

# Plot the line graph
sns.lineplot(x=ab.index, y=ab['mean'], marker='o', linewidth=4, color='darkblue')

# Title and labels with formatting
plt.title("Top 10 Years with the Highest Average Worldwide Collection", fontsize=18, weight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Worldwide Collection (in billions)", fontsize=14)

# Remove scientific notation from y-axis
plt.ticklabel_format(style="plain", axis='y')

# Add data labels for each point
for x, y in zip(ab.index, ab['mean']):
    plt.text(x, y, f'{y:,.2f}', ha='center', va='bottom', fontsize=12, weight='bold')

# Display the plot
plt.show()
```

```python
m = total["Domestic_percent"].mean()

# Generate a colormap
eda = total.query('Domestic_percent != "0"')\
    .groupby("year")["Domestic_percent"].mean()
colors = plt.cm.viridis(np.linspace(0, 1, len(eda)))

ax= eda.plot(kind="barh", x="year",y="Domestic_percent",color=colors)
# Generate a colormap
ax.set_xlabel("Average domestic percent collection in % every year")

ax.ticklabel_format(style="plain",axis="x")
for i in ax.containers:
    ax.bar_label(i, fmt='%.2f')
```

```python

# Filter and group data
eda = total.query('Foreign_percent != "0"')\
    .groupby("year")["Foreign_percent"].mean()
plt.figure(figsize=(14,5))
ax =sns.barplot(data= total,x="year",y="Foreign_percent",palette="muted")

# Set labels and title
ax.set_xlabel("Year")
ax.set_ylabel("Foreign(%)")
ax.set_title("Average Foreign percent collection by year")
plt.show()
```

```python


foreign = total.query('Foreign_percent != "0"')\
    .groupby("year")["Foreign_percent"].mean().reset_index()
domestic = total.query('Domestic_percent != "0"')\
    .groupby("year")["Domestic_percent"].mean().reset_index()

# Combine the data into one DataFrame with consistent column names
foreign['Type'] = 'Foreign'
domestic['Type'] = 'Domestic'
foreign.rename(columns={'Foreign_percent': 'Percent'}, inplace=True)
domestic.rename(columns={'Domestic_percent': 'Percent'}, inplace=True)
combined = pd.concat([foreign, domestic])

# Set the figure size
plt.figure(figsize=(16, 8))

# Plot using seaborn
ax = sns.barplot(data=combined, x="year", y="Percent", hue="Type", palette="viridis")

# Set labels and title
ax.set_xlabel("Year")
ax.set_ylabel("Average Percent Collection (%)")
ax.set_title("Average Foreign and Domestic Percent Collection by Year")

plt.legend(title='Percent Type')
plt.show()
```
